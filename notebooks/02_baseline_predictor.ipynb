# 02 Baseline Predictor

Compare simple non-neural baselines on a sliding-window dataset.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
SRC = ROOT / 'src'
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

from predictive_brain_lab.data import generate_sine_wave
from predictive_brain_lab.datasets import SlidingWindowDataset
from predictive_brain_lab.metrics import mean_absolute_error, mean_squared_error
from predictive_brain_lab.models import MovingAverageBaseline, PersistenceBaseline
from predictive_brain_lab.train import predict_baseline
```

```python
signal = generate_sine_wave(
    n_steps=400,
    frequency=2.0,
    amplitude=1.0,
    noise_std=0.05,
    sampling_rate=100.0,
    seed=7,
)
dataset = SlidingWindowDataset(signal, window_size=30)

y_true, y_pred_persistence = predict_baseline(PersistenceBaseline(), dataset)
_, y_pred_mean = predict_baseline(MovingAverageBaseline(), dataset)

print('Persistence MSE:', mean_squared_error(y_true, y_pred_persistence))
print('Persistence MAE:', mean_absolute_error(y_true, y_pred_persistence))
print('Moving average MSE:', mean_squared_error(y_true, y_pred_mean))
print('Moving average MAE:', mean_absolute_error(y_true, y_pred_mean))
```
